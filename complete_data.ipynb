## Analizar datos del s2

```python
import os
import json
import pandas as pd
from pprint import pprint
```

```python
# Directorio principal que contiene los subdirectorios de las entidades
directorio_principal = './PDN_S2'

# Lista para almacenar los datos de todos los archivos
datos = []

# Recorrer los subdirectorios
for entidad in os.listdir(directorio_principal):
    directorio_entidad = os.path.join(directorio_principal, entidad)
    if os.path.isdir(directorio_entidad):
        # Recorrer los archivos en el subdirectorio de la entidad
        for archivo in os.listdir(directorio_entidad):
            if archivo.endswith('.json'):
                ruta_archivo = os.path.join(directorio_entidad, archivo)
                with open(ruta_archivo, 'r', encoding='utf-8') as f:
                    datos_archivo = json.load(f)
                    datos.extend(datos_archivo)

# Crear un DataFrame con los datos
df = pd.DataFrame(datos)

# Mostrar el DataFrame
print(df)
```

```python
pprint(df.shape)
pprint(df.columns)
pprint(df.head())
pprint(df.tail())
pprint(df.info())
pprint(df.describe())
pprint(df.isnull().sum())
pprint(df.dtypes)
pprint(df.memory_usage(deep=True))
```
